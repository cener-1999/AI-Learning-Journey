# 隐藏层的激活函数

## Step Function 阶跃函数

```python
import numpy as np

def step_function(x: np.ndarray):
    y = x > 0
    return y.astype(np.int64)

x = np.array([1, 1, -2])
step_function(x)
```

```python
from matplotlib import pyplot as plt
from typing import Callable

def show_function(data, func: Callable):
    fig, ax = plt.subplots(figsize=(3, 2.7), layout='constrained')
    ax.plot(data, func(data))

x = np.arange(-5, 5, 0.1)
show_function(x, func=step_function)
```

## Sigmoid Function

```python
def sigmoid_function(x: np.ndarray):
    y = 1/ (np.exp(-x) + 1)
    return y

x = np.array([1, 1, -2])
sigmoid_function(x)
```

```python
x = np.arange(-5, 5, 0.1)
show_function(x, sigmoid_function)
```

## ReLU Function (Rectified Linear Unit)

```python
def relu_function(x: np.ndarray):
    return np.maximum(x, 0)

x = np.array([5, 5, -1])
relu_function(x)
```

```python
x = np.arange(-5, 5, 0.1)
show_function(x, relu_function)
```

# 输出层的激活函数

## Identity Function 恒等函数

```python
def identity_function(a: np.ndarray):
    return a
```

```python
x = np.arange(-5, 5, 0.1)
show_function(x, identity_function)
```

## softmax_function

**常用在分类问题的激活层中**

特征：
- 输出是0 ~ 1之间的实数
- 输出总和是1 -> **概率**

```python
def softmax_function(a: np.array):
    c = np.max(a)
    exp_a = np.exp(a+c)
    denominator = sum(exp_a)
    return exp_a / denominator

x = np.array([0, -10, -20])
softmax_function(x)
```
