```python
from langchain_openai import OpenAIEmbeddings
from settings import OPENAI_API_KEY
```

```python
embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
text = 'Hello World'
query_result = embeddings.embed_query(text)
query_result, len(query_result)
```

```python
words = ['cola', 'water', 'cry', 'smile']
doc_vectors = embeddings.embed_documents(words)
doc_vectors
```

```python
from scipy.spatial.distance import pdist, squareform
import numpy as np
import pandas as pd


print(len(doc_vectors))
print(len(doc_vectors[0]))

X = np.array(doc_vectors)

print(pdist(X))  # n维空间中观测值之间的成对距离
distance = squareform(pdist(X)) # Euclidean distances
distance
```

```python
df = pd.DataFrame(data=distance, index=words, columns=words)
df
```
